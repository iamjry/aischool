```python
import sqlite3
import numpy as np
import pandas as pd
import math
conn = sqlite3.connect('hackson.db')
sqlstr="SELECT 產品編號, 產品名稱, age15, age30, age50, age65, male, female, class1, class2, class3,prodname,prodname2,prodname3 FROM 中油生技產品;"
bioproduct=pd.read_sql(sqlstr,conn)
def manhattan_distance(x, y):   
    return sum(abs(a-b) for a,b in zip(x,y))
def update_product(sortedbioproduct,prodno):
  updatesql="update 中油生技產品 set prodno='%s', prodno2='%s', prodno3='%s', prodname='%s',  prodname2='%s', prodname3='%s' \
  where   產品編號='%s'" %(sortedbioproduct.iloc[0,0],sortedbioproduct.iloc[1,0],sortedbioproduct.iloc[2,0],\
                       sortedbioproduct.iloc[0,1],sortedbioproduct.iloc[1,1],sortedbioproduct.iloc[2,1],prodno)
  conn.execute(updatesql)
  conn.commit()
for i   in range(0,len(bioproduct) ):
  
  feature=bioproduct[['age15','age30','age50','age65','male','female','class1','class2','class3']]   
  for j   in range(0,len(bioproduct) ): 
      distance=manhattan_distance(feature.iloc[i],feature.iloc[j])
      bioproduct.loc[j,'distance']=distance
    
  sortebiodproduct=bioproduct.sort_values(by=['distance'],ascending=1)
  sortebiodproduct=sortebiodproduct[sortebiodproduct['產品編號']!=bioproduct.iloc[i,0]]
  #print(i)  
  update_product(sortebiodproduct,bioproduct.iloc[i,0])
  
```
